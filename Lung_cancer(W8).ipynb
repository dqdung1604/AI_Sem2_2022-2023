```python
import pandas as pd
url = 'lung_cancer_examples.csv'
data = pd.read_csv(url)
data.head()
```

```python
data = data.drop('Name', axis = 1)
data = data.drop('Surname', axis = 1)
data.head()
```

```python
from sklearn import datasets
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

x = data.loc[:, ['Age', 'Smokes', 'AreaQ', 'Alkhol']]
y = data['Result']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)

sc = StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
print(x_train, x_train_std)
```

```python
sc = StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
print(x_train, x_train_std)
```

```python
model = Perceptron(max_iter = 1000, eta0 = 0.1, random_state = 0)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(y_pred)
print(y_test)
print('Độ chính xác mô hình chưa scale:', accuracy_score(y_test, y_pred))
```
